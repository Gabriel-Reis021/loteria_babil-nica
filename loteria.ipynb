```python
import random

def get_input():
    while True:
        try:
            num_user = int(input("Entre com um n√∫mero: "))
        except ValueError:
            print("Valor inv√°lido: tente novamente")
            continue
        if 1 <= num_user <= 15:
            return num_user
        else:
            print("Valor inv√°lido! O valor deve ser entre 1 e 15")

def verify_numbers(n_sorteio, n_user):
    if n_user == n_sorteio:
        print("üéâüéâüéâ üí∏Parab√©ns! Voc√™ se tornou um milion√°rio!üí∏ üéâüéâüéâ")
        return True
    elif n_user > n_sorteio:
        print("N√∫mero muito alto. Tente novamente! ‚¨ÜÔ∏è")
        return False
    else:
        print("N√∫mero muito baixo. Tente um n√∫mero maior! ‚¨áÔ∏è")
        return False

jogar_novamente = True
while jogar_novamente:
    num_sorteio = random.randint(1, 15)
    for i in range(3):
        print(f"Tentativa {i+1} de 3")
        num_user = get_input()
        if verify_numbers(n_sorteio=num_sorteio, n_user=num_user):
            break
    else:
        print("Suas tentativas acabaram!")

    resp = input("Deseja jogar novamente? (S/N): ").strip().lower()
    if resp == "s":
        jogar_novamente = True
    else:
        jogar_novamente = False
```
